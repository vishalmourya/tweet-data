```python
raw_df=df.groupby(['user_id','user_screen_name', "tweet_text"])["retweets"].sum().reset_index()
raw_df
```

```python
raw_df['retweets'][:5].plot(kind='hist')
```

```python
sns.histplot(raw_df["retweets"][:5],bins=5)
```

```python
raw_df['retweets'].describe(percentiles=[0,1/5,2/5,3/5,4/5,1]).reset_index()
```

```python
raw_df.drop_duplicates(subset="retweets", keep=False, inplace=True)
```

```python
raw_df.shape
```

```python
sns.histplot(raw_df["retweets"][:10],bins=10)
```

```python
labels=["A(Less Data)", "B(Medium Data)","C(Large Data)","D(Maximum Data)"]
raw_df["quartile"] = pd.qcut(x=raw_df['retweets'], q=4,precision=0)
raw_df["quartile_labels"]= pd.qcut(x=raw_df["retweets"],q=4,labels=labels)
```

```python
raw_df["quartile_labels"].value_counts()
```
